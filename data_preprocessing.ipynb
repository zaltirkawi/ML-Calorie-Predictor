```python
# Splitting the nutrition column into separate columns
import ast
df['nutrition'] = df['nutrition'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
```

```python
# Column names based on information from dataset website
nutrition_columns = ['calories', 'total_fat', 'sugar', 'sodium', 'protein', 'saturated_fat', 'carbohydrates']

# Expanding the list of column names into separate columns
df[nutrition_columns] = df['nutrition'].apply(pd.Series)

# Dropping original nutrition column
df.drop(columns=['nutrition'], inplace=True)
```

```python
# Removing unnecessary features
df = df.drop(columns=['id', 'contributor_id', 'submitted', 'tags', 'steps', 'description','n_steps', 'minutes', 'n_ingredients'])
#df = df.drop(columns=['n_steps', 'minutes'])
```
